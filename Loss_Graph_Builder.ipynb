```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from google.colab import drive
```

```python
cnn_name = "ResCNN"

drive.mount('/content/drive')
file_name = "/content/drive/MyDrive/Papers/SELIC-Journal/Training-Performance/"+str(cnn_name)+".csv"
```

```python
df = pd.read_csv(file_name)
```

```python
y1 = df['valid_loss'].tolist()
y2 = df['train_loss'].tolist()
x = range(1, len(y2) + 1)


fig, ax1 = plt.subplots(figsize=(10, 6))

ax2 = ax1.twinx()
lns1 = ax1.plot(x, y1, 'b--', linewidth=2, label="Validation Loss")
lns2 = ax2.plot(x, y2, 'red', linewidth=2, label="Training Loss")


x_range = range(0, len(y2) + 1)
plt.xticks(np.arange(0, len(x)+1, 5))
plt.title(cnn_name, fontsize=18)

ax1.set_xlabel('Epochs', fontsize=14)
ax1.set_ylabel('Validation Loss', color='b', fontsize=14)
ax2.set_ylabel('Training Loss', color='red', fontsize=14)


# added these three lines for insert labels
lns = lns1+lns2
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs, loc=0, frameon=False, fontsize=14)

#plt.grid()
#plt.figure()
plt.savefig('/content/drive/MyDrive/Papers/SELIC-Journal/Training-Performance/resultados_dir/'+'graf_'+cnn_name+'.pdf')
#plt.show()
```
